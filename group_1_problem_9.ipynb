```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

# For value 0.2

```python
# In this code, we will show how to use K-mean clustering using sklearn...

# Lets first generate some 2-D training data for easier visualization...

# The task is to classify the data inteo K=4 clusters...

# We will use gaussian distributions to generate the data belonging to 4 classes. The means are located at (1,1), (-1,1), (1,-1), (-1,-1). 

# We will use different variances to test our system

# Remember that we have to shuffle the data when we input it to the Kmean algorithm. The algorithm can not know the labels as this an unsupervised learning technique

# Compare the EM to others... see "https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html"

number_of_clusters_K = 8
number_of_data_points_per_cluster = 10
number_of_data_point = number_of_clusters_K*number_of_data_points_per_cluster

data_1 = np.random.normal(loc=(1,0), scale = 0.2,   size=(number_of_data_points_per_cluster,2)) 
data_2 = np.random.normal(loc=(0.7,0.7), scale = 0.2,  size=(number_of_data_points_per_cluster,2)) 
data_3 = np.random.normal(loc=(0,1), scale = 0.2, size=(number_of_data_points_per_cluster,2)) 
data_4 = np.random.normal(loc=(-0.7,0.7), scale = 0.2,  size=(number_of_data_points_per_cluster,2)) 
data_5 = np.random.normal(loc=(-1,0), scale = 0.2,   size=(number_of_data_points_per_cluster,2)) 
data_6 = np.random.normal(loc=(-0.7,-0.7), scale = 0.2,  size=(number_of_data_points_per_cluster,2)) 
data_7 = np.random.normal(loc=(0,-1), scale = 0.2, size=(number_of_data_points_per_cluster,2)) 
data_8 = np.random.normal(loc=(0.7,-0.7), scale = 0.2,  size=(number_of_data_points_per_cluster,2)) 

# print("data_1 = ", data_1)
# print("data_2 = ", data_2)
# print("data_3 = ", data_3)
# print("data_4 = ", data_4)

# concatinate the data and shuffle
data = np.concatenate((data_1,data_2,data_3,data_4,data_5,data_6,data_7,data_8))

# shuffle the data: 
np.random.shuffle(data)
#print(data)
```

```python
# Now, lets call the K mean algorihtm and train using the first 20 data points...

kmeans = KMeans(n_clusters=8, random_state=0, algorithm = "auto").fit(data)

# printing the cluster centers: 
#print(kmeans.cluster_centers_)

# plot data points and the learned centers from kmeans...
for (x,y) in data: 
  plt.plot(x,y,'bx')
class_cnt = 0
for (x_c,y_c) in kmeans.cluster_centers_:
  plt.plot(x_c,y_c,'ro')
  plt.text(x_c,y_c, "Cluster "+str(class_cnt), fontsize=18)
  class_cnt = class_cnt +1

plt.show()
```

```python
# predicted labels for the training data
print(kmeans.labels_)

data_to_test = [0,0.1]
pred_label = kmeans.predict([data_to_test])
print("datapoint = ",data_to_test,"pred_label", pred_label)
```

```python
# print the predicted labels: 

for index in range(number_of_data_point):
  plt.plot(data[index][0],data[index][1], 'bx')
  plt.text(data[index][0],data[index][1], str(kmeans.labels_[index]))
class_cnt = 0
for (x_c,y_c) in kmeans.cluster_centers_:
  plt.plot(x_c,y_c,'ro')
  plt.text(x_c,y_c, "Cluster "+str(class_cnt), fontsize=18)
  class_cnt = class_cnt +1
plt.show()
```
